```python
import pandas as pd
```

```python
papers_file = "AIPapers.csv"
paper_aff_file = "AIPaperAff.csv"
aff_file = "Affiliations.txt"
aff_cols = ['AffiliationId', 'DisplayName', 'Iso3166Code']

papers = pd.read_csv(papers_file)
paper_aff = pd.read_csv(paper_aff_file)
aff = pd.read_csv(aff_file, delimiter='\t', names=aff_cols, usecols = [0, 3, 10])
```

```python
paper_aff_loc = paper_aff.merge(aff, on='AffiliationId', how='inner')
```

```python
paper_aff_loc
```

```python
paper_loc = paper_aff_loc[['PaperId', 'Iso3166Code']].drop_duplicates()
paper_loc
```

```python
publications_per_country = paper_loc.groupby("Iso3166Code").count().sort_values("PaperId", ascending=False).to_dict()['PaperId']
publications_per_country
```

```python
above_50k = dict()
above_10k = dict()
above_5k = dict()
above_1k = dict()
above_500 = dict()
below_500 = dict()

for k, v in publications_per_country.items():
    if v > 50000:
        above_50k[k.lower()]  = v
    elif v > 10000:
        above_10k[k.lower()]  = v
    elif v > 5000:
        above_5k[k.lower()]  = v
    elif v > 1000:
        above_1k[k.lower()]  = v
    elif v > 500:
        above_500[k.lower()]  = v
    else:
        below_500[k.lower()]  = v
```

```python
import pygal
from pygal.style import Style
```

```python
# create a world map
custom_style = Style(colors=("#007791", "#7CB9E8", "#1B4D3E", "#C9CC3F" ,"#FFBF00" , "#AA0000"))

worldmap =  pygal.maps.world.World(legend_at_bottom=True, style=custom_style)

# set the title of the map
worldmap.title = 'Number of publications in AI'
# adding the countries
worldmap.add('50,000 or more', above_50k)
worldmap.add('10,000 - 50,000', above_10k)
worldmap.add('5,000 - 10,000', above_5k)
worldmap.add('1,000 - 5,000', above_1k)
worldmap.add('500 - 1,000', above_500)
worldmap.add('1 - 500', below_500)
worldmap.show_legend = True
worldmap.human_readable = True
worldmap.render_to_png("AI_pub.png")
```
